Desarrolle un código que pida al usuario una palabra y muestre por pantalla si es un palíndromo.

Palíndromo: Palabra o expresión que es igual si se lee de izquierda a derecha que de derecha a izquierda.


```python
# Definición de la clase Cadenas
class Cadenas:
    def __init__(self, cad1):
        """Constructor de la clase que inicializa la variable cad1"""
        self.cad1 = cad1  # Se almacena la cadena ingresada en el atributo de instancia

    def Pal(self):
        #Método que verifica si la palabra almacenada en cad1 es un palíndromo"""
        cad1 = self.cad1  # Se obtiene la cadena de la instancia

        # Se inicializan variables
        c, i, nom, cad, x = 0, 0, '', '', ''  
        i = len(cad1)  # Se obtiene la longitud de la cadena
        nom = cad1.lower()  # Se convierte la cadena a minúsculas para evitar diferencias por mayúsculas

        # Inversión de la cadena usando un bucle while y un bucle for
        while i != c:  # Mientras la cantidad de caracteres invertidos no sea igual a la longitud de la cadena
            for x in nom:  # Se recorre cada caracter de la cadena original
                cad = x + cad  # Se construye la cadena invertida agregando cada caracter al inicio de 'cad'
                c = c + 1  # Se incrementa el contador para salir del bucle cuando se complete la inversión

            # Comparación entre la cadena original en minúsculas y su versión invertida
            if nom == cad:
                return str(cad1 + " Es Palindromo")  # Si son iguales, se retorna que la palabra es un palíndromo
            else:
                return str(cad1 + " No es Palindromo")  # Si no son iguales, se retorna que la palabra no es un palíndromo

# Solicita al usuario ingresar una palabra
cad1 = input('Dame una palabra: ')  

# Se crea un objeto de la clase Cadenas con la palabra ingresada
op1 = Cadenas(cad1)  

# Se ejecuta el método Pal para determinar si es un palíndromo y se muestra el resultado
print(op1.Pal())
#-------------------------------------------------------------
x = str(input("Ingrese una palabra: "))
if x == x[::-1]:
  print("Es un palindromo")
else:
  print("No es un palindromo")
```

Desarrolle un programa que solicite el ingreso de los siguientes datos de 5 trabajadores:
-	Nombre (solo un nombre)
-	Edad
-	Inasistencias en el mes 
-	Sueldo base 

Estos datos deben almacenarse en 4 diferentes listas.
Calcule el bono de gratificación que recibirá cada persona según las siguientes condiciones:

a)	Si la edad esta entre 20 y 30 el sueldo base tendrá una gratificación de un 15% sobre el sueldo total.
b)	Si la edad esta entre 31 y 40 el sueldo base tendrá una gratificación de un 30% sobre el sueldo total.
c)	Si la edad está por sobre los 41 años, el sueldo base tendrá una gratificación de un 40%.  
Al mismo tiempo agregue una penalización según las inasistencias en el mes, según el siguiente criterio:
a)	Si ha faltado entre 0 a 5 días, una penalización del 5% sobre el total (S. Base + Grat)
b)	Si ha faltado entre 6 a 10 días, una penalización del 10% sobre el total (S. Base + Grat)
c)	Si ha faltado entre 11 a 15 días, una penalización del 15% sobre el total (S. Base + Grat) y se informará que el trabajador debe ser despedido.
Al final del programa deberá imprimir por pantalla el nombre de cada persona, su edad, el sueldo base, el sueldo con la gratificación y algún descuento o condición de despido si los hubiese.


```python
# Listas para almacenar datos de los trabajadores
nombres = []
edades = []
inasistencias = []
sueldos_base = []

# Ingreso de datos de los 5 trabajadores
for i in range(5):
    nombre = input(f"Ingrese el nombre del trabajador {i+1}: ")
    edad = int(input(f"Ingrese la edad de {nombre}: "))
    inasistencia = int(input(f"Ingrese las inasistencias del mes de {nombre}: "))
    sueldo_base = float(input(f"Ingrese el sueldo base de {nombre}: "))

    nombres.append(nombre)# Agregar nombre a la lista
    edades.append(edad)
    inasistencias.append(inasistencia)
    sueldos_base.append(sueldo_base)

# Cálculo de gratificación y penalización
for i in range(5):
    sueldo_base = sueldos_base[i]

    # Calcular gratificación
    if 20 <= edades[i] <= 30:
        gratificacion = sueldo_base * 0.15
    elif 31 <= edades[i] <= 40:
        gratificacion = sueldo_base * 0.30
    elif edades[i] > 41:
        gratificacion = sueldo_base * 0.40
    else:
        gratificacion = 0  # Sin gratificación fuera de estos rangos

    sueldo_total = sueldo_base + gratificacion

    # Calcular penalización
    if 0 <= inasistencias[i] <= 5:
        penalizacion = sueldo_total * 0.05
    elif 6 <= inasistencias[i] <= 10:
        penalizacion = sueldo_total * 0.10
    elif 11 <= inasistencias[i] <= 15:
        penalizacion = sueldo_total * 0.15
        mensaje_despido = "¡El trabajador debe ser despedido!"
    else:
        penalizacion = 0
        mensaje_despido = ""

    sueldo_final = sueldo_total - penalizacion

    # Mostrar resultados
    print("\n--- Información del trabajador ---")
    print(f"Nombre: {nombres[i]}")
    print(f"Edad: {edades[i]}")
    print(f"Sueldo base: {sueldos_base[i]}")
    print(f"Sueldo con gratificación: {sueldo_total}")
    print(f"Descuento por inasistencias: {penalizacion}")
    print(f"Sueldo final: {sueldo_final}")
    if inasistencias[i] >= 11:
        print(mensaje_despido)
```

Calcula el factorial de un número dado usando un bucle. n! = n*(n-1)*...

```python
n = int(input("Ingrese un numero entero: "))
x = 1 # Inicializar x en 1 porque 1 es el valor identidad para la multiplicación
# Calcular el factorial de n
for i in range(1,n+1):
    x = x * i
print("El factorial de ", str(n), "es: ", str(x))
```

funciones predefinidas


```python
palabra = input("Ingrese una palabra: ")
longitud = len(palabra)
for i in range(longitud): #Recorre y/o cuenta cada letra de la palabra
    x = palabra[longitud - i - 1]
    print(x, end="") #Imprime la letra de la palabra en orden inverso
    # Esto seria con la palabra hola [4 - 0 - 1] = h, [4 - 1 - 1] = o, [4 - 2 - 1] = l, [4 - 3 - 1] = a
    if palabra == x:
        print("Es un palindromo")
    else:
        print("No es un palindromo")
   
```

/*
 * Escribe un programa que se encargue de comprobar si un número es o no primo.
 * Hecho esto, imprime los números primos entre 1 y 100.
 */

```python
def es_primo(numero):
    if numero <= 1:
        return False
    if numero == 2:
        return True
    if numero % 2 == 0:
        return False
    i = 3
    while i * i <= numero:
        if numero % i == 0:
            return False
        i += 2
    return True

print("Números primos entre 1 y 100:")
for numero in range(1, 101):
    if es_primo(numero):
        print(numero, end=" ")
print()
```
